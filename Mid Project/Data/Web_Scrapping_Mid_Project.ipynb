```python
#load libraries
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import numpy as np
import json
import pandas as pd
import csv
HEADERS ={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.0.0 Safari/537.36", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}
url='https://egywatch.com/'
response=requests.get(url,headers=HEADERS)
response
```

```python
# for pages from 1 to 53
watch_data=[]
rates={'1.00 stars':1,'2.00 stars':2,'3.00 stars':3,'4.00 stars':4,'5.00 stars':5}
rank=0
for i in range(1,54) :
    url='https://egywatch.com/collections/watches/men?page='+str(i)+'&sort_by=best-selling'
    response=requests.get(url,headers=HEADERS)
    print(response)
    print(f'for page {i}\n')
    soup = BeautifulSoup(response.content, 'html.parser')   
    products=soup.find_all('div',attrs={'class':'product-wrap'})
    print(f'number of products : {len(products)}\n')
    links=[]
    for p in products:
        link_filter=p.find('a',attrs={'class':'hidden-product-link'})
        link='https://egywatch.com/'+link_filter.get('href')
        links.append(link)
    
    for l in links :
        url=l
        response=requests.get(url,headers=HEADERS)
        print(response)
        print(l)
        print()
        soup = BeautifulSoup(response.content, 'html.parser')
        rank +=1
        print(f"for product {rank} in page {i} \n")
        try :
            brand_name_span=soup.find('span',attrs={'class':'vendor'}).find('a').get_text()
            brand_name=brand_name_span.strip()
        except :
            brand_name=np.nan
        try:    
            model_name_span=soup.find('h1',attrs={'class':'product_name title'}).get_text()
            model_name=model_name_span.strip()
        except :
            model_name=np.nan
        try :
            rate=rates[soup.find('span',attrs={'class':'jdgm-prev-badge__stars'}).get('aria-label')]
        except :
            rate=np.nan
        price_span=soup.find('span',attrs={'class':'money'}).get_text()
        price_span=price_span.split('EGP')[-1]
        price_span=price_span.replace(',','')
        price=price_span.replace('.00','')
        price=price.strip()
        if price != '0' :
            aval=True
        else :
            aval=False
        try:    
            reference_number_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__reference-number custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            reference_number=reference_number_span.strip('\n').strip()
        except :
            reference_number=np.nan
        try:    
            series_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__series custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            series=series_span.strip('\n').strip()
        except:
            series=np.nan
        try :
            movement_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__movement custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            movement=movement_span.strip('\n').strip()
        except:
            movement=np.nan
        try:    
            color_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__strap-color custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            color=color_span.strip('\n').strip()
        except:
            color=np.nan
        try:    
            material_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__strap-type custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            material=material_span.strip('\n').strip()
        except:
            material=np.nan
        try :    
            made_in_span=soup.find('ul',attrs={'class','tabs-content'}).find('div',attrs={'class':'custom-field custom-field__made-in custom-field__type--text'}).find('span',attrs={'class':'custom-field--value'}).get_text()
            made_in=made_in_span.strip('\n').strip()
        except :
            made_in=np.nan
        watch={'rank':rank,'brand_name':brand_name,'series':series,'model_name':model_name,'color':color,'material':material,
               'rate':rate,'price':price,'avaliability':aval,'made_in':made_in,
               'reference_number':reference_number,'movement':movement,'link':url    

              }
        print(watch)
        print()
        watch_data.append(watch)
        print("-------------------------------------------")    
```

```python
df=pd.DataFrame(watch_data)
df
```

```python
df.to_csv('top_100_best_sells_watch.csv',index=False)
```
